## 7.1 Negentropy is scale-invariant

The covariance matrix of $x$: $\Sigma = E[(x-E[x])(x-E[x])^{T}]$ 

The covariance matrix of $y$: $\Sigma_y = E[(Ax-E[Ax])(Ax-E[Ax])^{T}]$ where $y = Ax$ 

A is invariant for $E[]$ Therefore

\begin{eqnarray}
\Sigma_y &=& E[(Ax-AE[x])(Ax-AE[x])^{T}] \\
 &=& E[A(x-E[x])(A(x-E[x]))^{T}] \\
 &=& A \:E[(x-E[x])((x-E[x]))^{T}A^T] \\
\Sigma_y  &=& A \Sigma A^T
\end{eqnarray}

Then $H(y_{Gauss})=\frac{1}{2} \log{|\det{\Sigma_y}|}+ \frac{1}{2}(1+\log{2\pi})$

\begin{eqnarray}
H(y_{Gauss})&=&\frac{1}{2} \log{|\det(\Sigma_y)|}+ \frac{1}{2}(1+\log{2\pi}) \\
&=&\frac{1}{2} \log{|\det( A \Sigma A^T)|}+ \frac{1}{2}(1+\log{2\pi}) 
\end{eqnarray}

$|\det( A \Sigma A^T)|=|\det( A) \det( \Sigma ) \det(A^T)| $ 

$=|\det( A) \det( \Sigma ) \det(A)| $ 

$=\det( A)^2 |\det( \Sigma )| $ 

Therefore 

\begin{eqnarray}
& &\frac{1}{2} \log{|\det( A \Sigma A^T)|}+ \frac{1}{2}(1+\log{2\pi}) \\
&=& \frac{1}{2} \log{\det( A)^2 |\det( \Sigma )|}+ \frac{1}{2}(1+\log{2\pi}) \\
&=& \log{\det( A)}+ \frac{1}{2} \log{|\det( \Sigma )|}+ \frac{1}{2}(1+\log{2\pi}) \\
H(y_{Gauss})&=& \log{\det( A)}+ H(x_{Gauss}))
\end{eqnarray}

If we estimate a probability density is maltivariate gaussian

\begin{eqnarray}
p(x) = \frac{1}{(\sqrt{2\pi})^m \sqrt{|\det(\Sigma)|}} \exp(-\frac{1}{2} (x-E[x])^T \Sigma^{-1}(x-E[x]))
\end{eqnarray}

Then 

\begin{eqnarray}
p(Ax) &=& \frac{1}{(\sqrt{2\pi})^m \sqrt{\det( A)^2 |\det( \Sigma )|}} \exp(-\frac{1}{2} (Ax-E[Ax])^T (A \Sigma A^T)^{-1}(Ax-E[Ax])) \\
&=& \frac{1}{\det( A)} \frac{1}{(\sqrt{2\pi})^m \sqrt{|\det( \Sigma )|}} \exp(-\frac{1}{2} (Ax-E[Ax])^T (A \Sigma A^T)^{-1}(Ax-E[Ax])) \\
&=& \frac{1}{\det( A)} \frac{1}{(\sqrt{2\pi})^m \sqrt{|\det( \Sigma )|}} \exp(-\frac{1}{2} (x-E[x])^TA^T (A \Sigma A^T)^{-1}A(x-E[x])) \\
&=& \frac{1}{\det( A)} \frac{1}{(\sqrt{2\pi})^m \sqrt{|\det( \Sigma )|}} \exp(-\frac{1}{2} (x-E[x])^TA^T (A^{-T} \Sigma^{-1} A^{-1})A(x-E[x])) \\
&=& \frac{1}{\det( A)} \frac{1}{(\sqrt{2\pi})^m \sqrt{|\det( \Sigma )|}} \exp(-\frac{1}{2} (x-E[x])^T \Sigma^{-1} (x-E[x])) \\
p(Ax) &=& \frac{1}{\det( A)} p(x) \\
\end{eqnarray}

We have the differential entropy:

\begin{eqnarray}
H(x)= -\int p(x)\log{p(x)} dx \\
\end{eqnarray}

\begin{eqnarray}
H(Ax)&=& -\int \frac{1}{\det( A)} p(x) \log{\frac{1}{\det( A)} p(x)} dx \\
&=&  -\frac{1}{\det( A)} \int p(x) \log{\frac{1}{\det( A)} p(x)} dx \\
&=&  -\frac{1}{\det( A)} \int p(x)( \log{p(x)}-\log{\det( A)}) dx \\
&=&  -\frac{1}{\det( A)} \int p(x) \log{p(x)}- p(x)\log{\det( A)} dx \\
&=&  -\frac{1}{\det( A)} \int p(x) \log{p(x)} dx+\frac{1}{\det( A)} \int p(x)\log{\det( A)} dx \\
&=&  - \int p(x) \log{p(x)} dx+\log{\det( A)} \int p(x) dx \\
\end{eqnarray}

$ \int p(x) dx = 1$ Therefore

\begin{eqnarray}
&=&  - \int p(x) \log{p(x)} dx+\log{\det( A)} \\
H(Ax)&=& H(x) +\log{\det( A)}
\end{eqnarray}

The negentropy is defined as

\begin{eqnarray}
J(x) = H(x_{Gauss}) -H(x)
\end{eqnarray}

Then:

\begin{eqnarray}
J(Ax) &=& H(Ax_{Gauss}) -H(Ax) \\
&=& \log{\det( A)}+ H(x_{Gauss}) - (H(x) +\log{\det( A)}) \\
&=& H(x_{Gauss}) - H(x) = J(x)
\end{eqnarray}

Therefore the negentropy is invariant w.r.t. invertible linear transformations.

If $A=aI$ where $a \in \mathbb{R}$, $I$ is identical matrix $I \in \mathbb{R}^{n \times n} $ it is analog that the negentropy is scale invariant.

## 7.2 fastICA vs. Infomax 

Apply fastICA to the two soundfiles data set (once again) of problem sheet 6 and compare runtime and robustness w.r.t. the mixing matrix A with the Infomax-based ICA-algorithm. Use the following setup for the latter algorithm: natural gradient, Bell-Sejnowski amplitude normalization, learning rate schedule "0 = 0:01, "t+1 = 0:9999"t.

```python
unzip = zipfile.ZipFile("sounds.zip", 'r')          
fl = unzip.namelist()                     
s1 = unzip.open(fl[0])
s2 = unzip.open(fl[1])
dat1 = pd.read_table(s1,header=None,delim_whitespace=True)
dat2 = pd.read_table(s2,header=None,delim_whitespace=True)
sig1 = np.array(dat1)
sig2 = np.array(dat2)
S= np.append(sig1 ,sig2,1)
A =np.array( [[ 0.63610688,0.46237165], [ 0.95995455, 0.17711994]])
W = np.array([[ 0.58870338, 0.42567703],[ 0.66824231,0.62259454]])
X = np.dot(A,S.T)
nn = len(X[0])
rmidx = np.random.choice(nn,nn,replace=False)
XX=X[:,rmidx]
XX = XX -np.mean(XX,1).reshape(2,1)
```

```python
def sigmoid(y):
    return 1/(1+np.exp(-y))

def natGrad(X, W, eps=0.01,l=0.9999):
    n = len(X)
    p=len(X[0])
    dw = np.array(W)
    ww = np.array(W)
    rmidx = np.random.choice(p,p,replace=False)
    XX=X[:,rmidx]
    # natural gradient  
    for i in range(p):
        y = np.dot(ww,XX[:,i].reshape(n,1))
        sig = 1 -2*sigmoid(y)
        dw= ww + np.dot(np.dot(sig,XX[:,i].reshape(1,n)),np.dot(ww.T,ww))
        ww=ww+eps*dw
        eps=eps*l
    return ww

def regGrad(XX, W, eps=0.01,l=0.9999):
    n = len(XX)
    p=len(XX[0])
    dw = np.array(W)
    ww = np.array(W)
    #“regular” gradient online
    for i in range(p):
        if(np.linalg.det(ww)==0.):
            ww = ww+np.eye(n)*0.0000000001
        invW= np.linalg.solve(ww,np.eye(n))
        y = np.dot(ww,XX[:,i].reshape(n,1))
        sig = 1 -2*sigmoid(y)
        dw =invW.T +np.dot(sig,XX[:,i].reshape(1,n))
        ww=ww+eps*dw
        eps=eps*l
    return ww
```

```python
tm =time.time()
wr=regGrad(XX, W, eps=0.01,l=0.9999)
sr = np.dot(wr,X)
print("Time of regulär gradient: ",time.time()-tm)
```

```python
tm =time.time()
wn=natGrad(XX, W, eps=0.01,l=0.9999)
sn = np.dot(wn,X)
print("Time of  natural gradient: ",time.time()-tm)
```

```python
tm =time.time()
ica = FastICA(n_components=2)
S_ = ica.fit_transform(X.T)  # Reconstruct signals
A_ = ica.mixing_  # Get estimated mixing matrix
print("Time of  fastICA: ",time.time()-tm)
```

(d) Perform PCA on the same set of patches, plot the the principal components (ordered by decreasing eigenvalue) as in (c) and compare them with the independent features.

```python
class myPCA():
    def __init__(self, X):
        # ...
        C = np.cov(X)
        D, U = np.linalg.eigh(C)
        idx = np.argsort(-D)
        self.U = U[:,idx]
        self.D = D[idx]
        
    def project(self, X, m):
        # ...
        U = self.U
        M = U[:,:m] 
        Z = np.dot(M.T,X)
        return Z
    
    def denoise(self, X, m):
        Z = self.project(X,m)
        U = self.U
        Y = np.dot(U[:,:m],Z)
        Y = Y + np.mean(X,axis=1).reshape(len(Y),1)
        return Y

```
